```python
fruit_protein = {
    "apple": 0.3,
    "banana": 1.3,
    "cherry": 1.0,
    "date": 2.5,
    "elderberry": 0.7,
    "fig": 1.2,
    "grape": 0.6,
    "honeydew": 0.8,
    "kiwi": 1.1
}
```

```python
s2 = pd.Series(fruit_protein, name="Protein")
```

```python
s2
```

```python
s2>1
```

```python
s2[s2>1]
```

```python
[(s2>1) & (s2<2)]
```

```python
[(fruit, protein) for fruit, protein in fruit_protein.items() if protein > 1]
```

```python
# not operation 
s2[~(s2>1)]
```
