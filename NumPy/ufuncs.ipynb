Universal function -> ufuncs

```python
import numpy as np
```

```python
arr = np.array([1, 4, 9, 16])
```

```python
print(np.sqrt(arr))  # Output: [1. 2. 3. 4.]
```

```python
print(np.exp([1, 2]))
```

```python
angles = np.array([0, np.pi, np.pi/2])
print(np.sin(angles))  # Output: [ 0.  0.  1.]
```
