```python
import numpy as np
```

```python
arr = np.array([1, 2, 3, 4, 5, 6])
reshaped = arr.reshape(2, 3)
print(reshaped)
print(reshaped.shape)
```

```python
reshaped2 = reshaped.reshape(3, 2)
reshaped2
```

```python
ravel = reshaped2.ravel()
ravel
```

```python
ravel[0] = 99
print(ravel)
print(reshaped2)
```

```python
flat = reshaped2.flatten()
flat
```

```python
flat[0] = 1
flat
```
