```python
import numpy as np
```

```python
arr = np.array([[[1, 2], [3, 4]]])
arr.shape
```

```python
swap = np.swapaxes(arr, 0, 2)
swap.shape
```

```python
# Concatenation:

a = np.array([1, 2])
b = np.array([3, 4])
```

```python
combine = np.concatenate((a, b))
combine
```

```python
arr1 = np.array([[1, 2], [3, 4]])
arr2 = np.array([[5, 6], [7, 8]])
```

```python
print(np.vstack((arr1, arr2)))
```

```python
print(np.hstack((arr1, arr2)))
```
