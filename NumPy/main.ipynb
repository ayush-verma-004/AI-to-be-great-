# Dimensions and shape of an array

```python
print(arr.ndim)
print(arr.shape)
```

```python
arr = np.array(34)
print(arr.ndim)
print(arr.shape)
```

```python
arr = np.array([[1, 2, 3], [4, 5, 6]])
print(arr.ndim)
print(arr.shape)
```
