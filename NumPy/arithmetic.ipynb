```python
import numpy as np
```

```python
a = np.array([1, 2, 3])
b = np.array([4, 5, 6])
```

```python
print(a+b)  # Output: [5 7 9]
```

```python
print(a-b)  # Output: [-3 -3 -3]
```

```python
print(a/b) # Output: [0.25 0.4  0.5 ]
```

```python
print(a*b) # Output: [ 4 10 18 ]
```

```python
print(a//b) # Output: [0 0 0 ]
```

```python
print(b//a) # Output: [4 2 2 ]
```

```python
print(b%a) # Output: [0 1 0 ]
```

```python
print(a**2)
```
